# ZoneOpt.jl Model Training Demo
This notebook demonstrates how to train a Neural ODE for hockey zone entry trajectories using ZoneOpt.jl. It covers loading processed data, defining the model, running training, and visualizing loss curves.

```python
# 1. Import Libraries and Setup
using DiffEqFlux, OrdinaryDiffEq, Plots, CSV, DataFrames
include("../src/neural_ode_framework.jl") # Assumes ZoneOpt.jl logic is in src/
using ZoneOpt
```

```python
# 2. Basic Package Initialization
processed_sequences = CSV.read("../processed/neural_ode_sequences_long.csv", DataFrame)
sequence_index = CSV.read("../processed/neural_ode_sequence_index.csv", DataFrame)
```

```python
# 3. Demonstrating Core Functions
# Define and train the Neural ODE model
model, loss_history = ZoneOpt.train_neural_ode(processed_sequences, sequence_index)
plot(loss_history, title="Training Loss Curve", xlabel="Epoch", ylabel="Loss")
```

```python
# 4. Advanced Features and Configurations
# Custom training parameters and validation split
params = Dict(:epochs => 50, :batch_size => 32, :learning_rate => 1e-3)
model, loss_history = ZoneOpt.train_neural_ode(processed_sequences, sequence_index; params...)
```

```python
# 5. Handling Exceptions and Warnings
try
    model, loss_history = ZoneOpt.train_neural_ode(processed_sequences, sequence_index)
catch e
    @warn "Training failed: $e"
end
```
